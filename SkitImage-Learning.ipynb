```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage import data,io
import glob
%matplotlib inline
```

`io.imread_collection()`读取一组图片，里面可以使用通配符进行加载图片
>使用skimage读入的直接就是ndarray格式
读入的是三通道RGB（height，width，channel）,使用的时候用[0][1]维度就可以，后面的通道数3不用管，处理方法和普通方法一样，最后整理数据集的时候用转置transpose

```python
path=os.path.join(os.getcwd(),'Train/*.bmp')
print (path)
imgs=io.imread_collection(path)#使用读取一组图片
for img in imgs:
    plt.imshow(img)
print(img.shape)
```

图片裁剪
