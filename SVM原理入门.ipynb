# 相关概念
线性分类器：线性分类器是分类器中的一种，就是判定分类结果的根据是通过特征的现行组合得到的，不能通过特征的非线性运算结果作为判定依据。
线性分类器：线性分类器是分类器中的一种，就是判定分类结果的是通过特征的线性组合得到的，不能通过特征的肺相信运算结果作为判定依据。

## 线性分类器的起源
在实际应用中，我们往往遇到这样的问题：给定一些数据点，他们分别属于两个不同的类，现在要找到一个线性分类器把这些数据分成两类。怎么分呢？把整个空间切割开来，直线左边的便属于类别-1，直线右边的点属于类别1。在二维空间中，用一条直线就把空间分割开来了，在高维空间中(n>3),需要n-1维的超平面将空间分割开。<br>
常见的线性分类器有感知器模型和逻辑回归。上一届举出的例子是感知器模型，直接给你分好类。有时候，我们除了要知道分类器对于新数据的分类结果，还希望知道分类器对于这次分类的成功概率，这就是逻辑回归。<br>
逻辑回归（虽然称为回归，但却是一个分算法）将现行分类器的超平面方程计算结果通过logistic函数从正负无穷映射到0-1，这样的映射结果就可以认为是分类器将x(数据样本)判定为类别的概率

## SVM（support vector machines）
是一种二分类模型，他的基本模型是定义在特征空间上间隔最大的线性分类器，建个最大使它有别于干之家；支持向量机还包括核技巧，这使它称为实质上的费线性分类器。支持向量机的学习策略就是间隔最大化，可形式化为一个求解凸二次规划(convex quadratic programming)的问题，也等价于正则化的合页损失函数最小化问题。支持向量机的学习算法就是求解凸二次规划的最优化算法。



## SVM求解思路
可以使用$y_i(wx+b)$来表示分类正确性和确信度，这就是函数间隔(functional margin)，这一概念在下面被用作为超平面分类正确的判断条件。<br>
**高维空间中点到超平面的距离：**$\frac {\hat \gamma }{||w||}$
因此我们需要最大化这个间隔距离，同时满足分类正确,最大化
$$max \frac{\hat \gamma}{||w||}$$
$$y_i(wx+b)\ge 1$$
函数间隔$\hat \gamma$的取值不影响最优化问题的解。事实上，函数间隔的变化不影响优化问题，这样可以取$\hat \gamma=1$最大化$\frac {1}{||w||}$与最小化$1\2||w||^2$是等价的，于是可以得到下面的支持向量优化问题
$$min_{w,b}\qquad\frac {1}{2}||w||^2$$
$$s.t.\qquad y_i(w\cdot x_i+b)$$
