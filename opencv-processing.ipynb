```python
import cv2
import matplotlib.pyplot as plt
```

`cv2.imwrite(fileName, image, [cv2.IMWRITE_JPEG_QUALITY, 100])`表示存储图像质量

```python
img = cv2.imread('pytorch-yolo3-python3/output/0.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst = cv2.GaussianBlur(img, (3,3), 1)
plt.imshow(dst)
```

```python
img = cv2.imread('pytorch-yolo3-python3/output/0.jpg')
b,g,r = cv2.split(img)  # split之后 得到三个通道的数据
bH = cv2.equalizeHist(b)
gH = cv2.equalizeHist(g)
rH = cv2.equalizeHist(r)
result = cv2.cvtColor(cv2.merge((bH,gH,rH)), cv2.COLOR_BGR2RGB)  # 将三个通道合成在一起
plt.imshow(result)
```

```python
img = cv2.imread('mashiro.jpg',3)
cv2.imshow('mashiro',img)
b,g,r = cv2.split(img)  # split之后 得到三个通道的数据
bH = cv2.equalizeHist(b)
gH = cv2.equalizeHist(g)
rH = cv2.equalizeHist(r)
result = cv2.merge((bH,gH,rH))  # 将三个通道合成在一起
cv2.imshow('mashiro1',result)
```

## 图像的滤波
`Image`对象继承了`ImageFilter`的filter方法,但是滤波的方法需要从ImageFilter中调用

```python
pil_img = Image.open('pytorch-yolo3-python3/output/0.jpg')
pil_img = pil_img.filter(ImageFilter.GaussianBlur)#返回图像对象,注意必须要用返回值哦
# print(pil_img)
pil_img.show()
```
