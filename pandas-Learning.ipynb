## 数据索引
1.选择一个单独的列,浙江返回一个Series,等同于`df.A`
2.通过[]进行选择,这将对`行`进行切片

可以通过`DataFrame.loc()`实现行列索引
`DataFrame.loc[]`实现切片索引,注意不是调用函数(),而是使用[]进行索引
* 单个标签,有些index可能是str类型,索引的时候直接使用相应的字符标志就可以,这个标识符可能是行标也可以是列标
* 一个label的list 例如['a', 'b', 'c']
* 切片索引,例如['a':'c',1:3]
* 如果得到了index的一个seq,可以使用loc索引出所有的行
>使用单个index label进行索引的时候,如果在行列中出现同样的label,以行先出现的为准<br>
使用`[start:end]`切片索引的时候首尾都包括,这一点与np有头无尾不一样<br>
多列索引必须使用双层括号的方式**`df[['A','B']], df.loc[:,['A','B','C']]`**!!!!!


给loc传递条件判断的参数其实是一个bool的list[True, True, ....] length相当对于df的行数

## Join
类似SQl类型合并,这些方法表现出很好的性能
`pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, )`通过键拼接列,用于通过一个或多个键将两个数据集的行连接起来，类似于 SQL 中的 JOIN。该函数的典型应用场景是，针对同一个主键存在两张包含不同字段的表，现在我们想把他们整合到一张表里。在此典型情况下，结果集的行数并没有增加，列数则为两个元数据的列数和减去连接键的数量。<br>
**on:**:依赖连接键值<br>
**how:**'inner' 参数指的是当左右两个对象中存在不重合的键时，取结果的方式：inner 代表交集；outer 代表并集；left 和 right 分别为取一边。
>默认以重叠的列名当做连接键

````python
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,)
````

````python
left = pd.DataFrame({'key': ['foo', 'foo'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [3, 4]})
pd.merge(left, right)
````

## `groupby()`
这是一个分组工具,根据字段默认给一些元素分组,分组后不会对源数据结构更改,相当于返回一个按某columns进行索引划分出几个组,
$shape_{origin}=(A, B)$,经过groupby后变为$shape_{after}=(C,D,E)$,C为以group(`key`),划分出组的数量,数据量没有发生改变,只是说排列方式发生了变化

````python
df = {'city':['Beijing', 'Shanghai', 'Guangzhou', 'Shenzhen','Hangzhou', 'Chongqing', 'Chengdu'],
       'year':[2016, 2016, 2015, 2017, 2016, 2016, 2017],
       'population':[2100, 2300, 1000, 700, 500, 500, 600]}
frame = pd.DataFrame(df)
frame.groupby('year')
````

通过groupby()返回`DataFrameGroupBy`类型对象,这个对象并非不能直接使用行列索引得到group的分组,但是具有迭代功能<br>
frame.groupby('year')[行或列] xxx这样是错误的行为<br>
分组的目的主要在于统计其他column的信息,因此对groupby后面使用相关的统计函数会返回以groupby('label')的分组信息作为行索引,列还是原来的其他column,每一条的的数据是相应的统计数据.<br>
打个比方,原来label中有数据总共有这三种类型[2015, 2016, 2017],对label进行了groupby之后,总的分组变成了三类,使用统计信息功能后,会返回一个行index为[2015. 2016, 2017],列为之前column统计值的DataFrame<br>
`.sum()`对应组内求和`.count()`统计不同元素出现的频率`.mean()`组内元素求均值<br>
`DataFrameGroupBy.agg(['sum', 'mean', 'count'])`返回一个具有MultiLabel的DataFrame,一次性可以计算多个统计特性
`DataFrameGroupBy.get_group(group_index)`可以获取每个分组的信息,返回值为`DataFrame`,

````python
frame.groupby('year').count()
````

````python
frame.groupby('year').mean()#无法求均值的column会被自动舍弃
````

````python
frame.groupby('year').agg(['sum', 'count'])#一次可以展示多个统计结果
````

上述groupby是整个DataFrame进行处理的,我们也可以抓出单独的几条columns按另一column进行groupby,例如:
