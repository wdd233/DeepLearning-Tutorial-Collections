# matplotlib
绘图和可视化，主要应用于2D图表的绘制
下面是一个简单的例子

```python
%matplotlib inline
import numpy as np
from numpy.random import randn
import matplotlib
import matplotlib.pyplot as plt
plt.plot(np.arange(10))
```

## Figure and Subplot
matplotlib 的图像都位于Figure对象中，可以用plt.figure创建一个新的图表
matplotlib中的figure还支持一种matlab的编号方式
不能通过空的Figure绘图，必有创建一个或者多个plot才行


```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)#意思是：图像应该是2×2的，当前选中的是第一个
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
fig#fig包含四个子区间图
```

## axis
设置坐标轴
plt.axis((x1,x2,y1,y2))将图的横坐标范围(x1-x2),纵坐标(y1-y2)
>注意参数要以元组的形式输入

```python
plt.axis((1,10,20,30))
```

```python
plt.plot(randn(50).cumsum(),'k--')
ax2.scatter(np.arange(30),np.arange(30)+3*randn(30))
```

出现了一个更为方便的方法plt.subplots,它可以新建一个Figure，并返回一个含有已经创建的subplot对象的numpy数组

```python
fig,axes=plt.subplots(2,3)
axes
```

***

```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
```

## 如何区别不同类别的颜色呢？
在进行分类操作中，最终结果分为不同的类，我们只用一个plt.scatter()来进行绘图，但是如何区别不同类别的颜色?<br/>
我们可以使用c可以输入一个numpy()数组，因此只需要x[(1类)(2类)(3类)]对应的label[(1类)(2类)(3类)]

```python
a0=np.random.normal(0,1,(1,10))
a1=np.random.normal(1,1,(1,10))
b0=np.random.normal(2,1,(1,10))
b1=np.random.normal(4,1,(1,10))
y0=np.ones(10)
y1=3*np.ones(10)
x0=np.concatenate((a0,b0),axis=1)#注意，如果a0是二维的的话要用
x1=np.concatenate((a1,b1),axis=1)
y=np.concatenate((y0,y1),axis=0)#y作为label，同时也可以代表不同的颜色
plt.scatter(x0[0],x1[0],c=y)
```
